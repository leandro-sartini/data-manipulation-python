# Iterando por dataframes.

### Podemos iterar por Dataframes também, de uma maneira um pouco diferente, onde o primeiro parâmetro passado no for é sua coluna.

```python
for column in merged:
    print(column)
```

# E se a gente quiser iterar por linhas? Existe uma função do pandas chamada iterrows().

```python
for index,row in merged.iterrows():
    print(row)
```

## Ai podemos criar loops monstrinhos com isso!

# NÃO FAÇAM O EXEMPLO ABAIXO, pois pode travar seu computador!

```python
for column in merged:
    for index,row in merged.iterrows():
        print(merged[column])
```
